```python
import dtale
import dtale.app as dtale_app
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
import pandas as pd
tracks=pd.read_csv(io.BytesIO(uploaded['tracks.csv']))
tracks['name']=tracks['name'].fillna('unknown')
tracks['release_date']=pd.to_datetime(tracks['release_date'])  ## Convert String to Date/Time
tracks['Year']=pd.DatetimeIndex(tracks['release_date']).year ## Extract Year
tracks['Month']=pd.DatetimeIndex(tracks['release_date']).month ## Month
tracks['artists']=tracks['artists'].str.replace("[","")
tracks['artists']=tracks['artists'].str.replace("]","")
tracks['artists']=tracks['artists'].str.replace("'","")
tracks['id_artists']=tracks['id_artists'].str.replace("[","")
tracks['id_artists']=tracks['id_artists'].str.replace("]","")
tracks['id_artists']=tracks['id_artists'].str.replace("'","")
tracks=tracks[tracks['Year']!=1900]  ## Dropping rows with year equal to 1900
tracks['Minutes']=round(tracks['duration_ms']/60000,2)
tracks_Lata_M=tracks[tracks['artists']=="Lata Mangeshkar"]
tracks_Lata_M.head(5).sort_values(by='popularity',ascending=False)
```

```python
dtale.app.USE_NGROK=True ## Used for Google Collab
dtale.show(tracks_Lata_M)
```
